## 5. 
**a. Maak een nieuwe dataset met per unieke waarde voor de variabele Handelsbenaming het
aantal keren dat deze voorkomt. Welke benaming komt het vaakste voor? **

```python
new_df = pd.DataFrame()
for name in honda_df['Handelsbenaming'].unique() :
    new_df['Handelsbenaming'] = name
    new_df['Count'] = honda_df['Handelsbenaming'].value_counts()

print(new_df)

```
